```python
import socket
import pyautogui
import pyperclip

UDP_IP = "0.0.0.0" 
UDP_PORT = 5005
sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# FIX: Is line se "Address already in use" wala error khatam ho jayega
sock.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)

sock.bind((UDP_IP, UDP_PORT))
print("PC-2 Ready: Waiting for Teleport/Paste...")

while True:
    try:
        data, addr = sock.recvfrom(4096)
        msg = data.decode().split("|")
        command = msg[0]
        
        if command == "MOVE":
            x, y = float(msg[1]), float(msg[2])
            pyautogui.moveTo(x, y)
        elif command == "PASTE":
            pyperclip.copy(msg[1])
            pyautogui.hotkey('ctrl', 'v')
            print("Pasted from PC-1!")
    except:
        continue
```
